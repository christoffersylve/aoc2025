# Part 1

```python
with open("input.txt") as input:
  lines = input.readlines()

edges = []
for line in lines:
  u,vs = line.strip().split(":")
  vs = vs.strip().split()
  for v in vs:
    edges.append((u,v))

part1 = 0

stack = [x for x in edges if x[0] == "you"]

while stack:
  u = stack.pop()[1]
  vs = [e[1] for e in edges if e[0] == u]
  for v in vs:
    if v == "out":
      part1 += 1
    else:
      stack.append((u,v))
    
print("Part 1 Answer",part1)
```
