# Part 1

```python

import numpy as np
import math

nbr_to_connect = 1000

with open("input.txt") as input: 
  lines = input.readlines()

junction_boxes = [[int(y) for y in x.strip().split(",")] for x in lines]

connections = []
for i in range(len(junction_boxes)):
  for j in range(i,len(junction_boxes)):
    if i != j:
      connections.append((i,j,np.linalg.norm(np.array(junction_boxes[i])-np.array(junction_boxes[j]))))

connections = sorted(connections, key=lambda x: x[2])
circuits = []

for index, box in enumerate(junction_boxes):
  circuits.append({index})
for link in connections[:nbr_to_connect]:

  temp_set_a, temp_set_b = {}, {}

  first_index = next(
    (i for i, box in enumerate(circuits) if {link[0]}.issubset(box)),
    None
  )

  if first_index != None:
    temp_set_a = circuits.pop(first_index)

  second_index = next(
    (i for i, box in enumerate(circuits) if {link[1]}.issubset(box)),
    None
  )

  if second_index != None:
    temp_set_b = circuits.pop(second_index)

  circuits.append(temp_set_a.union(temp_set_b))
  

part1 = math.prod(sorted([len(x) for x in circuits])[len(circuits)-3:])
print("Part 1 Answer",part1)
```
