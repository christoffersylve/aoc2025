# Part 1

```python

with open('input.txt') as input:
  data = [line.strip() for line in input]
  ranges = [list(map(int,r.split("-"))) for r in data[:data.index("")]]
  ids = [int(r) for r in data[data.index("")+1:]]

fresh_ingredients = 0

for id in ids:
  for range in ranges:
    if range[0] <= id and id <= range[1]:
      fresh_ingredients += 1 
      break

print("Part 1 Answer:",fresh_ingredients)  
```

# Part 2

```python

with open('input.txt') as input:
  data = [line.strip() for line in input]
  ranges = sorted([list(map(int,r.split("-"))) for r in data[:data.index("")]], key=lambda r: r[0])

unique_ranges = 0
current_range = []

for index,range in enumerate(ranges[:len(ranges)]):
  current_range = range if not current_range else current_range
  if current_range[1] >= range[0]:
    if current_range[1] <= range[1]:
      current_range[1] = range[1]
  else:
    unique_ranges += current_range[1]-current_range[0]+1
    current_range = range

#Adding the final range
unique_ranges += current_range[1]-current_range[0]+1

print("Part 2 Answer:",unique_ranges)
```
