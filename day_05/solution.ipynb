# Part 1

```python

with open('input.txt') as input:
  data = [line.strip() for line in input]
  ranges = [list(map(int,r.split("-"))) for r in data[:data.index("")]]
  ids = [int(r) for r in data[data.index("")+1:]]

fresh_ingredients = 0

for id in ids:
  for range in ranges:
    if range[0] <= id and id <= range[1]:
      fresh_ingredients += 1 
      break

print("Part 1 Answer:",fresh_ingredients)  
```
