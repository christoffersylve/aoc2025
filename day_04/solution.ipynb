# Part 1

```python
input = []

with open("input.txt") as lines:
  for line in lines:
    input.append(list(line.strip()))

nbr_accessible_rolls = 0

for r, row in enumerate(input):
  for c, col in enumerate(row):
    nbr_adjecent_rolls = -1
    temp = []
    if col == "@":
      for i in range(max(r-1,0),min(r+2,len(input))):
        for j in range(max(c-1,0),min(c+2,len(row))):
          temp.append(input[i][j])
          if input[i][j] == "@":
            nbr_adjecent_rolls += 1
      if nbr_adjecent_rolls < 4:
        nbr_accessible_rolls += 1

print("Part 1 Answer:",nbr_accessible_rolls)
```

# Part 2

```python
input = []

with open("input.txt") as lines:
  for line in lines:
    input.append(list(line.strip()))

nbr_removed_rolls = 0

roll_removed=True

while roll_removed:
  roll_removed=False
  for r, row in enumerate(input):
    for c, col in enumerate(row):
      nbr_adjecent_rolls = -1
      temp = []
      if col == "@":
        for i in range(max(r-1,0),min(r+2,len(input))):
          for j in range(max(c-1,0),min(c+2,len(row))):
            temp.append(input[i][j])
            if input[i][j] == "@":
              nbr_adjecent_rolls += 1
        if nbr_adjecent_rolls < 4:
          input[r][c] = "."
          roll_removed = True
          nbr_removed_rolls += 1

print("Part 2 Answer:",nbr_removed_rolls)
```
