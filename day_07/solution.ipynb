# Part 1

```python
rows = []

with open("input.txt") as input:
  for line in input: 
    rows.append(list(line.strip()))

part1 = 0

for r, row in enumerate(rows):
  for c, col in enumerate(row):
    if rows[r-1][c] == "|":
      if col == "^":
        rows[r][c-1] = "|"
        rows[r][c+1] = "|"
        part1 += 1
        
      else:
        rows[r][c] = "|"
   
print(part1)
```

# Part 2

```python
rows = []

with open("input.txt") as input:
  for line in input: 
    rows.append(list(line.strip()))

nbr_paths = {}
nbr_paths[(1,rows[1].index("|"))] = 1

part1 = 0

for r, row in enumerate(rows):
  for c, col in enumerate(row):
    if rows[r-1][c] == "|":
      if col == "^":
        rows[r][c-1] = "|"
        rows[r][c+1] = "|"
        # Check to the left
        if (r,c-1) in nbr_paths.keys():
          nbr_paths[(r,c-1)] += nbr_paths[(r-1,c)]
        else:
          nbr_paths[(r,c-1)] = nbr_paths[(r-1,c)]
        # Check to the right
        if (r,c+1) in nbr_paths.keys():
          nbr_paths[(r,c+1)] += nbr_paths[(r-1,c)]
        else:
          nbr_paths[(r,c+1)] = nbr_paths[(r-1,c)]

      else:
        rows[r][c] = "|"
        if (r,c) in nbr_paths.keys():
          nbr_paths[(r,c)] += nbr_paths[(r-1,c)]
        else:
          nbr_paths[(r,c)] = nbr_paths[(r-1,c)]
  
keys = [t for t in nbr_paths.keys() if t[0] == len(rows)-1]

part2 = 0
for key in keys:
  part2 += nbr_paths[key]

print("Part 2 Answer:",part2)
```
