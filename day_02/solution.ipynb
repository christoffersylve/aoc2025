# Part 1

```python
ranges = []
invalid_ids = 0

with open('input.txt') as input:
  for raw_input in input:
    raw_inputs = raw_input.split(",")
  for r in raw_inputs:
    start = r.split("-")[0]
    end = r.split("-")[1]
    ranges.append((int(start),int(end)))

for span in ranges:
  for i in range(span[0],span[1]+1):
    id = str(i)
    if id[:len(id)//2] == id[len(id)//2:]:
      invalid_ids += i
    
print("Part 1 Answer:",invalid_ids)
```

# Part 2

```python
ranges = []
invalid_ids = 0

with open('input.txt') as input:
  ranges = [(int(start), int(end)) 
    for line in input 
    for part in line.strip().split(",") 
    for start, end in [part.split("-")]]

for span in ranges:
  for i in range(span[0],span[1]+1):
    id = str(i)
    for j in range(1,len(id)):
      sub_strings = [id[k:k+j] for k in range(0,len(id),j)]
      if all(x == sub_strings[0] for x in sub_strings):
        invalid_ids += i
        break
    
print("Part 2 Answer:",invalid_ids)
```
