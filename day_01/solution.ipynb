# Part 1

```python

acctual_password = 0
current_pos = 50

with open("input.txt") as input:
  for line in input:
    direction = line[:1]
    nbr_clicks = int(line[1:])
    if direction == "R":
      current_pos = (current_pos + nbr_clicks) % 100
    else:
      current_pos = (current_pos - nbr_clicks) % 100
    if current_pos == 0:
      acctual_password += 1

print("Part 1 Answer:",acctual_password)
```

# Part 2

```python

acctual_password = 0
current_pos = 50

with open("input.txt") as input:
  for line in input:
    direction = line[:1]
    nbr_clicks = int(line[1:])
    if direction == "R":
      acctual_password += nbr_clicks // 100
      if (current_pos + (nbr_clicks%100)) > 100:
        acctual_password += 1
      current_pos = (current_pos + nbr_clicks) % 100
    else:
      acctual_password += nbr_clicks // 100
      if (current_pos - (nbr_clicks%100)) < 0 and current_pos != 0: # Edge case moving dial left from zero
        acctual_password += 1 
      current_pos = (current_pos - nbr_clicks) % 100
    if current_pos == 0:
      acctual_password += 1

print("Part 2 Answer:",acctual_password)
```
