# Part 1

```python
input = []
nbr_batteries = 2

with open("input.txt") as lines:
  for line in lines:
    input.append([int(x) for x in list(line.strip())])
  
jolts = 0

for row_nbr, row in enumerate(input):
  stack = []
  for i, num in enumerate(row):
    batteries_left_in_bank = len(row)-i
    while stack and stack[-1] < num:
      if ((len(stack)-1) + batteries_left_in_bank) >= nbr_batteries:
        stack.pop()
      else:
        break
    if len(stack) < nbr_batteries: 
      stack.append(num)

  jolt = 0
  for i, num in enumerate(stack):
    jolt += num * (10 ** (len(stack)-i-1))
  jolts += jolt

print("Part 1 Answer:",jolts)
```
